```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('C:/Users/DELL/batsman_season_record.csv')
```

```python
df.info()
```

# Vertical bargraph

```python
plt.bar(np.arange(df.shape[0]),df['2016'])
plt.xticks(np.arange(df.shape[0]),df['batsman'],rotation=40)
plt.show()
```

# Horizontal bargraph 

```python
plt.barh(np.arange(df.shape[0]),df['2017'],color='orange')
plt.yticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

# Multiple bargraph

```python

plt.bar(np.arange(df.shape[0])-.3,df['2015'],width=.3,color='orange')
plt.bar(np.arange(df.shape[0]),df['2017'],width=.3,color='yellow')
plt.bar(np.arange(df.shape[0])+.3,df['2016'],width=.3,color='green')
plt.xticks(np.arange(df.shape[0]),df['batsman'],rotation=40)
plt.show()
```
